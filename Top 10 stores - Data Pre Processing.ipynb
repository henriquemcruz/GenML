```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from datetime import datetime
```

```python
#Read the dataset
#dataset = pd.read_csv('train_top10.csv')
#data = pd.DataFrame(dataset)
holiday_events = pd.read_csv('holidays_events.csv')
items = pd.read_csv('items.csv')
oil = pd.read_csv('oil.csv')
stores = pd.read_csv('stores.csv')
transactions = pd.read_csv('transactions.csv')
```

```python
#Summary of number of transacations per day
trans_per_day = transactions.groupby(['date']).sum()
trans_per_day = trans_per_day.drop('store_nbr',1)
trans_per_day = pd.DataFrame(trans_per_day)
```

```python
#Transform transactions dataframe to have date as datetime in the index
transactions['date'] = pd.to_datetime(transactions['date'])
transactions.set_index('date', inplace=True)
```

```python
#Check monthly seasonality on transactions - Aggregation of transactions and ploting them
transactions_month = transactions.resample('M').sum()
plt.figure(figsize=(20,10))
plt.plot(transactions_month['transactions'])
```

```python
#Check weekly seasonality on transactions - Aggregation of transactions and ploting them
transactions_week = transactions.resample('W').sum()
plt.figure(figsize=(20,10))
plt.plot(transactions_week['transactions'])
```

```python
#Check weekly seasonality on transactions - Aggregation of transactions and ploting them
transactions_day = transactions.resample('D').sum()
plt.figure(figsize=(20,10))
plt.plot(transactions_day['transactions'])
```
